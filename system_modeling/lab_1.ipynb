# Модель по экономике: Простая модель совокупного спроса и предложения (AD-AS)

## 1. Описание модели

Модель совокупного спроса и предложения (AD-AS) используется для анализа
макроэкономических явлений, таких как экономический рост, инфляция и безработица. Она
описывает взаимодействие между совокупным спросом (AD) и совокупным предложением
(AS) в экономике.

*Спрос* — это запрос фактического или потенциального покупателя, потребителя на
приобретение товара по имеющимся у него средствам, которые предназначены для этой
покупки. Спрос отражает, с одной стороны, потребность покупателя в некоторых товарах
или услугах, желание приобрести эти товары или услуги в определенном количестве и, с
другой стороны, возможность оплатить покупку по цене, находящейся в пределах
«доступного» диапазона.

Вместе с этими обобщёнными определениями спрос характеризуется рядом свойств и
количественных параметров, из которых, прежде всего, следует выделить объём или
величину спроса.

С позиций количественного измерения спрос на товар, понимается как объём спроса,
означает количество данного товара, которое покупатели (потребители) желают, готовы и
имеют денежную возможность приобрести за некоторый период в определённом ценовом
диапазоне.

*Величина спроса* — количество товара или услуг определённого вида и качества, которое
покупатель желает купить по определённой цене за определённый период. Величина спроса
зависит от доходов покупателей, цен на товары и услуги, цен на товары-субституты и
комплементарные блага, ожиданий покупателей, их вкусов и предпочтений.
Закон спроса — величина (объём) спроса уменьшается по мере увеличения цены товара.
Математически это означает, что между величиной спроса и ценой существует обратная
зависимость. То есть повышение цены вызывает понижение величины спроса, снижение же
цены вызывает повышение величины спроса.

Неценовые факторы, влияющие на спрос:

* Уровень доходов в обществе;
* Размеры рынка;
* Мода, сезонность;
* Наличие товаров-субститутов (заменителей);
* Инфляционные ожидания.

Природа закона спроса не сложная. Если у покупателя есть определенная сумма денег на
приобретение данного товара, то он сможет купить тем меньше товара, чем больше цена и
наоборот. Конечно, реальная картина намного сложнее, так как покупатель может
привлечь дополнительные средства, купить вместо данного товара другой — товарсубститут. 

**Закон предложения** — при прочих неизменных факторах величина (объём) предложения
увеличивается по мере увеличения цены на товар.

Рост величины предложения на товар при увеличении его цены обусловлен в общем случае
тем обстоятельством, что при неизменных издержках на единицу товара с увеличением
цены растёт прибыль и производителю (продавцу) становится выгодным продать больше
товара. Реальная картина на рынке сложнее этой простой схемы, но выраженная в ней
тенденция имеет место.

Неценовые факторы, влияющие на предложение:

1. Наличие товаров-заменителей.
2. Наличие товаров-комплементов (дополняющих).
3. Уровень технологий.
4. Объём и доступность ресурсов.
5. Налоги и дотации.
6. Природные условия.
7. Ожидания (инфляционные, социально-политические).
8. Размеры рынка.

---

### Основные составляющие модели (AD-AS):

1. Совокупный спрос (AD): Отражает общий уровень расходов в экономике. Зависит от уровня
цен (P) и объема выпуска (Y):

$AD = C + I + G + (X - M)$

C: Потребительские расходы.
I: Инвестиции.
G: Государственные расходы.
X - M: Чистый экспорт.

2. Совокупное предложение (AS): Отражает общий объем выпуска в экономике, который фирмы
готовы произвести при определенном уровне цен. Зависит от:
- Потенциального выпуска (Y*),
- Коэффициента чувствительности предложения к изменению цен (k),
- Уровня цен (P).

$AS = Y* + k(P - Pe)$

Pe: Ожидаемый уровень цен.

3. Равновесие: Устанавливается, когда AD = AS.

### Запустим базовую программу 

````python
def simulate_ad_as(C, I, G, X, M, Y_star, k, P_e, P, dt, T):
 t = 0 # Начальное время
 results = [] # Список для хранения результатов
 print("Time\tPrice Level\tAD\tAS")
 while t < T:
    # Вычисление совокупного спроса
    AD = C + I + G + (X - M)
    # Вычисление совокупного предложения
    AS = Y_star + k * (P - P_e)
    
    # Корректировка уровня цен
    if AD > AS:
        P += 0.01 * (AD - AS) # Рост уровня цен
    else:
        P -= 0.01 * (AS - AD) # Снижение уровня цен
        
    # Сохранение результатов
    results.append((t, P, AD, AS))
    print(f"{t:.2f}\t{P:.2f}\t{AD:.2f}\t{AS:.2f}")
    # Увеличение времени
    t += dt
    
 return results


# Инициализация параметров
C = 100 # Потребительские расходы
I = 50 # Инвестиции
G = 70 # Государственные расходы
X = 30 # Экспорт
M = 20 # Импорт
Y_star = 200 # Потенциальный выпуск
k = 0.5 # Чувствительность предложения
P_e = 100 # Ожидаемый уровень цен
P = 100 # Начальный уровень цен
dt = 0.1 # Шаг времени
T = 50 # Общее время моделирования
# Запуск симуляции
simulate_ad_as(C, I, G, X, M, Y_star, k, P_e, P, dt, T)
````

### 1. Базовый эксперимент:

- Параметры: C = 100, I = 50, G = 70, X = 30, M = 20, Y* = 200, k = 0.5, Pe = 100.
- Цель: Наблюдать изменения уровня цен и динамику спроса и предложения в экономике.

````python
C, I, G, X, M = 100, 50, 70, 30, 20
Y_star, k, Pe = 200, 0.5, 100

def AD(P) -> float:
    return (C + I + G + (X - M)) - 0.5 * P  # AD = 230 - 0.5P

def AS(P) -> float:
    return Y_star + k * (P - Pe) # AS = 200 + 0.5(P - 100)

P_eq = 80  # Решение уравнения 230 - 0.5P = 200 + 0.5(P - 100)
Y_eq = AD(P_eq)

P_range = np.linspace(50, 150, 100)
Y_AD = [AD(p) for p in P_range]
Y_AS = [AS(p) for p in P_range]

plt.figure(figsize=(12, 7))
plt.plot(Y_AD, 
         P_range, 
         lw=2.5, 
         color='#1f77b4', 
         label='AD: $230 - 0.5P$',
         )

plt.plot(Y_AS, 
         P_range, 
         lw=2.5, 
         color='#ff7f0e', 
         label='AS: $200 + 0.5(P - 100)$',
         )
plt.scatter(Y_eq, 
            P_eq, 
            s=100, 
            color='red',
            zorder=5, 
            label=f'Равновесие (Y={Y_eq}, P={P_eq})'
            )

plt.title('Равновесие в модели AD-AS', fontsize=16, pad=20)
plt.xlabel('Реальный выпуск (Y)', fontsize=12)
plt.ylabel('Уровень цен (P)', fontsize=12)
plt.xticks(np.arange(150, 250, 10))
plt.yticks(np.arange(50, 160, 10))
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(frameon=True, loc='upper right', fontsize=10)
plt.annotate('Точка равновесия:\nP=80, Y=190', 
             xy=(Y_eq, P_eq), 
             xytext=(180, 60),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=10)

plt.show()
````

````python
from ipywidgets import interact, FloatSlider

def calculate_AD(P, C, I, G, X, M):
    """Рассчитывает совокупный спрос"""
    return (C + I + G + (X - M)) - 0.5 * P

def calculate_AS(P, Y_star, k, Pe):
    """Рассчитывает совокупное предложение"""
    return Y_star + k * (P - Pe)

def find_equilibrium(params_AD, params_AS):
    """Находит точку равновесия"""
    P = np.linspace(0, 200, 1000)
    AD_vals = calculate_AD(P, **params_AD)
    AS_vals = calculate_AS(P, **params_AS)
    idx = np.argwhere(np.diff(np.sign(AD_vals - AS_vals)))[0][0]
    return P[idx], AD_vals[idx]

def plot_ad_as(C, I, G, X, M, Y_star, k, Pe):
    plt.figure(figsize=(12, 7))
    
    params_AD = {'C': C, 'I': I, 'G': G, 'X': X, 'M': M}
    params_AS = {'Y_star': Y_star, 'k': k, 'Pe': Pe}
    
    P_range = np.linspace(50, 150, 100)
    Y_AD = [calculate_AD(p, **params_AD) for p in P_range]
    Y_AS = [calculate_AS(p, **params_AS) for p in P_range]
    
    try:
        P_eq, Y_eq = find_equilibrium(params_AD, params_AS)
    except:
        P_eq, Y_eq = np.nan, np.nan
    
    plt.plot(Y_AD, P_range, lw=2.5, color='blue', 
            label=f'AD = {sum(params_AD.values())} - 0.5P')
    plt.plot(Y_AS, P_range, lw=2.5, color='orange',
            label=f'AS = {Y_star} + {k}(P - {Pe})')
    
    if not np.isnan(P_eq):
        plt.scatter(Y_eq, P_eq, s=100, color='red', zorder=5,
                   label=f'Равновесие (Y={Y_eq:.1f}, P={P_eq:.1f})')

    plt.title('Динамическая модель AD-AS\n', fontsize=14)
    plt.xlabel('Реальный выпуск (Y)', fontsize=12)
    plt.ylabel('Уровень цен (P)', fontsize=12)
    plt.xlim(150, 250)
    plt.ylim(50, 150)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(loc='upper left')
    
    explanations = [
        "Эффекты изменений:",
        f"→ Потребление (C): {C}",
        f"→ Инвестиции (I): {I}",
        f"→ Госрасходы (G): {G}",
        f"→ Чистый экспорт: {X-M}",
        f"→ Потенциальный выпуск: {Y_star}",
        f"→ Эластичность AS: {k}",
        f"→ Ожидаемые цены: {Pe}"
    ]
    
    plt.annotate('\n'.join(explanations), 
                xy=(0.65, 0.65), 
                xycoords='axes fraction',
                fontsize=9, 
                bbox=dict(facecolor='white', alpha=0.9))
    
    plt.show()

interact(plot_ad_as,
         C=FloatSlider(value=100, min=50, max=150, description='C (Потребление)'),
         I=FloatSlider(value=50, min=0, max=100, description='I (Инвестиции)'),
         G=FloatSlider(value=70, min=0, max=150, description='G (Госрасходы)'),
         X=FloatSlider(value=30, min=0, max=50, description='X (Экспорт)'),
         M=FloatSlider(value=20, min=0, max=50, description='M (Импорт)'),
         Y_star=FloatSlider(value=200, min=150, max=250, description='Y* (Потенциал)'),
         k=FloatSlider(value=0.5, min=0.1, max=2.0, description='k (Эластичность)'),
         Pe=FloatSlider(value=100, min=80, max=120, description='Pe (Ожидания)'));
````

**Почему выпуск ниже Y*?**

````python
C, I, X, M = 100, 50, 30, 20
Y_star, k, Pe = 200, 0.5, 100
P_range = np.linspace(50, 150, 100)

def calculate_AD(G):
    """Рассчитывает кривую AD при разных G"""
    return [ (C + I + G + (X - M)) - 0.5*p for p in P_range ]

def calculate_AS():
    """Рассчитывает кривую AS"""
    return [ Y_star + k*(p - Pe) for p in P_range ]

def find_equilibrium(G):
    """Находит точку равновесия"""
    AD_vals = calculate_AD(G)
    AS_vals = calculate_AS()
    idx = np.argmin(np.abs(np.array(AD_vals) - np.array(AS_vals)))
    return P_range[idx], AD_vals[idx]

def plot_g_effect(G=70):
    plt.figure(figsize=(12, 7))
    
    Y_AD = calculate_AD(G)
    Y_AS = calculate_AS()
    
    P_eq, Y_eq = find_equilibrium(G)
    
    plt.plot(Y_AD, P_range, lw=3, color='blue', 
            label=f'AD = {C+I+G+(X-M)} - 0.5P (G={G})')
    plt.plot(Y_AS, P_range, lw=2, color='orange', 
            label=f'AS = 200 + 0.5(P - 100)')
    
    plt.scatter(Y_eq, P_eq, s=100, color='red', zorder=5,
               label=f'Равновесие (Y={Y_eq:.1f}, P={P_eq:.1f})')

    plt.annotate(f'Эффект увеличения G:\n+1 единица G → +{0.5/(1-0.5):.1f} к Y (мультипликатор)',
                xy=(0.65, 0.75), 
                xycoords='axes fraction',
                fontsize=10,
                bbox=dict(facecolor='white', alpha=0.9))
    
    plt.title('Динамика фискальной политики: Изменение госрасходов\n', fontsize=14)
    plt.xlabel('Реальный выпуск (Y)', fontsize=12)
    plt.ylabel('Уровень цен (P)', fontsize=12)
    plt.xlim(150, 260)
    plt.ylim(50, 150)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    
    plt.show()

interact(plot_g_effect,
         G=FloatSlider(value=70, min=50, max=150, step=5, 
                      description='Госрасходы (G)',
                      readout_format='.0f'));
````

### Интерпретация:

**Механизм стимулирования:**

Государство, увеличивая расходы (например, на инфраструктуру или соцпрограммы), напрямую повышает совокупный спрос. Это создает мультипликативный эффект: рост доходов → увеличение потребления → дальнейший рост AD.

**Инфляционные риски:**
Рост цен (P=80 → 110) объясняется:

* Усилением давления спроса на ограниченные ресурсы,

* Ростом издержек производства (например, зарплаты).

**Долгосрочные последствия:**

Со временем фирмы пересмотрят инфляционные ожидания (Pe вырастет), что сдвинет AS влево. Это вернет Y к Y*=200, но при более высоком уровне цен (например, P=120).

````python
C, I, G, X, M = 100, 50, 70, 30, 20
Y_star, k = 200, 0.5
P_range = np.linspace(50, 150, 100)

def calculate_AD():
    """Рассчитывает кривую AD"""
    return [ (C + I + G + (X - M)) - 0.5*p for p in P_range ]

def calculate_AS(Pe):
    """Рассчитывает кривую AS при разных ожиданиях цен"""
    return [ Y_star + k*(p - Pe) for p in P_range ]

def find_equilibrium(Pe):
    """Находит точку равновесия"""
    AD_vals = calculate_AD()
    AS_vals = calculate_AS(Pe)
    idx = np.argmin(np.abs(np.array(AD_vals) - np.array(AS_vals)))
    return P_range[idx], AD_vals[idx]

def plot_pe_effect(Pe=100):
    plt.figure(figsize=(12, 7))
    
    Y_AD = calculate_AD()
    Y_AS = calculate_AS(Pe)
    
    P_eq, Y_eq = find_equilibrium(Pe)
    
    plt.plot(Y_AD, P_range, lw=2, color='blue', 
            label=f'AD: 230 - 0.5P (фиксирован)')
    plt.plot(Y_AS, P_range, lw=3, color='orange',
            label=f'AS = 200 + 0.5(P - {Pe})')
    
    plt.scatter(Y_eq, P_eq, s=100, color='red', zorder=5,
               label=f'Равновесие (Y={Y_eq:.1f}, P={P_eq:.1f})')
    
    explanation_text = [
        "Эффект изменения ожиданий:",
        f"1. Pe↑ → AS сдвигается ВЛЕВО",
        f"2. Новое равновесие:",
        f"   - Цены: {P_eq:.1f} (+{P_eq-80:.1f} к базе)",
        f"   - Выпуск: {Y_eq:.1f} ({Y_eq-190:.1f} к базе)",
        f"3. Эффект стагфляции:",
        "   Рост цен + падение выпуска"
    ]
    
    plt.annotate('\n'.join(explanation_text), 
                xy=(0.65, 0.65), 
                xycoords='axes fraction',
                fontsize=10,
                bbox=dict(facecolor='white', alpha=0.9))
    
    plt.title('Шок предложения: Изменение инфляционных ожиданий\n', fontsize=14)
    plt.xlabel('Реальный выпуск (Y)', fontsize=12)
    plt.ylabel('Уровень цен (P)', fontsize=12)
    plt.xlim(150, 250)
    plt.ylim(50, 150)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    
    plt.show()

interact(plot_pe_effect,
         Pe=FloatSlider(value=100, min=80, max=120, step=5,
                       description='Ожидания цен (Pe)',
                       readout_format='.0f'));
````

## Выводы по работе с моделью AD-AS

### 1. Основные закономерности модели

* **Равновесие определяется пересечением AD и AS.**

    * Любые изменения в компонентах AD (C, I, G, X-M) или параметрах AS (Y*, k, Pe) смещают равновесие.

    * Пример: Увеличение G сдвигает AD вправо → рост Y и P (Эксперимент 2).

* **Ценовые и неценовые факторы:**

    * Цена (P) влияет на величину спроса/предложения в краткосрочном периоде.

    * Неценовые факторы (технологии, ожидания, налоги) определяют сдвиги кривых.

### 2. Роль ожиданий в экономике

* **Инфляционные ожидания (Pe):**

    * Рост Pe сдвигает AS влево → стагфляция (Y↓, P↑) (Эксперимент 3).

    * Ожидания могут стать самореализующимися: если фирмы ждут роста цен, они повышают P, что провоцирует реальную инфляцию.

* **Управление ожиданиями — ключевой инструмент для Центральных банков. Пример:**

    * Якорение инфляционных ожиданий через прозрачную монетарную политику.

### 3. Эффективность экономической политики

* **Стимулирование спроса (фискальная политика):**

    * Эффективно при наличии свободных ресурсов (Y < Y*).

    * При перегреве экономики (Y > Y*) ведет к инфляции (Эксперимент 2).

* **Шоки предложения:**

    * Требуют мер, направленных на восстановление AS: инвестиции в технологии, снижение налогов для бизнеса (Эксперимент 3).

* **Гибкость предложения (k):**

    * Высокий k усиливает реакцию экономики на шоки → резкие колебания Y и P (Эксперимент 5).

    * Низкий k замедляет адаптацию, но смягчает краткосрочные колебания.

### 4. Риски и вызовы

* **Дефляционная спираль:**

    * Падение инвестиций (I↓) → снижение AD → дефляция → дальнейшее сокращение расходов (Эксперимент 4).

    * Пример: Великая депрессия 1930-х годов.

* **Стагфляция:**

    * Сочетание стагнации (Y↓) и инфляции (P↑) требует комплексных мер: подавление инфляции + стимулирование предложения (Эксперимент 3).

### 5. Ограничения модели AD-AS

* **Упрощение реальности:**

    * Не учитывает денежный рынок и ставки процента (влияние монетарной политики).

    * Предполагает статичность ожиданий (Pe фиксирован), хотя в реальности они адаптивны.

* **Краткосрочный фокус:**

    * Долгосрочная кривая AS (LRAS) вертикальна при Y*, но в модели AS обычно рассматривается как восходящая.

### 6. Практические рекомендации

* **Для борьбы с рецессией:**

    * Стимулировать AD через G↑ или снижение налогов.

    * Поддерживать доверие инвесторов для роста I.

* **Для сдерживания инфляции:**

    * Ужесточение монетарной политики (повышение ставок).

    * Контроль инфляционных ожиданий через коммуникацию.

* **Для устойчивого роста:**

    * Инвестиции в технологии и образование (рост Y*).

    * Гибкость рынков (труда, товаров) для повышения эластичности AS.

### Итог:

**Модель AD-AS — мощный инструмент для анализа макроэкономической динамики, но ее применение требует учета:**

* Взаимодействия краткосрочных и долгосрочных факторов,

* Роли ожиданий и институтов,

* Ограниченности упрощенных предположений.
Пример использования: Анализ последствий COVID-19: шок предложения (разрыв цепочек) + шок спроса (падение C и I) → необходимость комбинированной политики.
